```python
import pandas as pd
```

```python
# Importar data de las reacciones adversas
reacciones_adversas = pd.read_csv('2019VAERSDATA.csv', encoding='ISO-8859-1')
```

```python
# Columnas de las reacciones adversas
reacciones_adversas.columns
```

```python
# Muestra las reacciones adversas
reacciones_adversas = reacciones_adversas.set_index('VAERS_ID')
reacciones_adversas.head(5)
```

```python
# Descripción de las reacciones adversas
reacciones_adversas.describe()
```

```python
# Importar vacunas
vacunas = pd.read_csv('2019VAERSVAX.csv', encoding='ISO-8859-1')
vacunas = vacunas.set_index('VAERS_ID')
vacunas.head(3)
```

```python
# Descripción de la data
vacunas.describe()
```

```python
# Filtrar sólo la vacuna en interes: 
vacunas = vacunas[vacunas.VAX_TYPE == 'DTAPIPVHIB']
vacunas.head(3)
```

```python
# Filtrando datos de reacciones adversas segun la vacuna Pentavalente
pentavalente_reacciones = reacciones_adversas[reacciones_adversas.index.isin(vacunas.index.values)]
pentavalente_reacciones.describe()
```

```python
# Agregando datos de las vacunas a las reacciones adversas
vacunas_reacciones = pentavalente_reacciones.join(vacunas)
vacunas_reacciones.head(3)
```

```python
vacunas_reacciones.info()
```
